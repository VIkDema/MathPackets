```python
import numpy as np
import scipy.signal as signal 
import matplotlib.pyplot as plt
import IPython
from scipy.io import wavfile 
from more_itertools import windowed
from pydub import AudioSegment
import soundfile as sf
from pydub import AudioSegment
import tempfile
import os
import librosa
%matplotlib inline 
plt.rcParams['figure.figsize'] = [20, 10]
```

Считываем данные и преобразуем стерио в моно
Строим амплитудный спектр для каждого трека.

```python
def load_audio(filepath):
    mp3_audio = AudioSegment.from_file(filepath, format="mp3")
    data=mp3_audio.get_array_of_samples()
    x=[]
    for i in range(0,len(data),2):
        x.append((data[i]+data[i+1])/2)
    return x
```

```python
N=256#размерность фурье
def amlitude_spectrum(data,N):
    x=librosa.stft(data,n_fft=N,win_length=int(N/2))
    amplitude_spectrum = np.abs(x)
    res=[]
    for i in range(len(amplitude_spectrum)):
        res.append(np.mean(amplitude_spectrum[i]))
    return res
```

```python
count_of_data=124
amplitude_spectrums=[]
for i in range(1,count_of_data+1):
    print(i,end=' ')
    if i <10:
        amplitude_spectrums.append(amlitude_spectrum(np.array(load_audio('0'+str(i)+'.mp3')),N))
    else:
        amplitude_spectrums.append(amlitude_spectrum(np.array(load_audio(str(i)+'.mp3')),N))
```

```python
plt.plot(amplitude_spectrums[0])
```

Для использования метода KMeans были выбраны следующие параметры:
    init='k-means++' - таким образом изначальные центройды будут выбраны не случайным образом, а по принятым данным, что увеличит точность результата.
    n_init и max_init
    
Для метода k_means важна размерность векторов признаков, так как при работе алгоритма вычисляется декартово расстояние между векторами, так для вектора отличающиеся только по одному компоненту расстояние может отличаться в разы, хотя при этом принадлежать одному классу.

Например,
x=[$A_1,A_2,...,A_{i-1},100000$] и y = [$A_1,A_2,...,A_{i-1},1]$

Декартово расстояние отличичается значительно, хотя при большо i скорее всего можно сказать, что данные векторы принадлежат одному классу.

Поэтому нашей задачей является правильно уменьшить размерность векторов.
Способы которые были применены для данной выборки данных:

1)Амплитудный спектр 

Для каждого кадра был сформирован средний спектр, что позволило в разы уменьшить разрядность, мы можем это заметить на графики выше, как меняются данные спектры.

2) Попробуем воспользоваться методом главных компонент и сравнить с метод k средних без использования PCA.

```python
for i in amplitude_spectrums:
    print(len(i))
print(len(amplitude_spectrums),len(amplitude_spectrums[0]))
print(type(amplitude_spectrums),type(amplitude_spectrums[0]),type(amplitude_spectrums[0][0]))
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5,init='k-means++').fit(amplitude_spectrums)
print("Классы",kmeans.labels_,sep=' : ')
print("Компактность кластеров",kmeans.inertia_,sep=' : ')
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)

pca_data = pca.fit(amplitude_spectrums).transform(amplitude_spectrums)

kmeans = KMeans(n_clusters=5,init='k-means++').fit(pca_data)

print("Классы",kmeans.labels_,sep=' : ')
print("Компактность кластеров",kmeans.inertia_,sep=' : ')

plt.scatter(pca_data[:,0],pca_data[:,1],c=kmeans.labels_,s=100)

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c=[0,1,2,3,4],marker="v")
plt.show()
```

```python
import csv 
def get_number(number):
    if number>-1 and number<10:
        return "0"+str(number)
    else:
        return str(number)
    
myData = [[get_number(i),kmeans.labels_[i-1]]for i in range(1,len(kmeans.labels_)+1)] 
myFile = open('result.csv', 'w') 
with myFile: 
    writer = csv.writer(myFile) 
    writer.writerows(myData) 
```
